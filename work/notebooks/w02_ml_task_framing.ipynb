## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

we predict how much a page's traffic is changing, and we got that number ourselves from two real time periods already in the data — not from a shortcut label.

```python
scoped = df[(df["avg_position"] > 0) & (df["impressions_prev_30d"] > 0)].copy()
```

```python
scoped
```

```python
scoped["impr_change_pct"] = (scoped["impressions_last_30d"] - scoped["impressions_prev_30d"]) / scoped["impressions_prev_30d"]
```

```python
scoped["impr_change_pct"].describe()
```

```python
print(f"Share with ANY decline (<0%): {(scoped['impr_change_pct'] < 0).mean():.1%}")
```

```python
print(f"Share with a >10% decline: {(scoped['impr_change_pct'] < -0.10).mean():.1%}")
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

A simple rule based on one obvious factor (word count) scores 0% precision — it's no better than picking pages at random. That's because decline isn't driven by any single factor alone; it comes from a combination of weak signals (search volume, competition, staleness, etc. — remember, we checked these earlier and each had near-zero correlation on its own). Only a model that can weigh multiple signals together has a chance of finding the real decliners.

```python
ranked = scoped.sort_values("impr_change_pct")
```

```python
ranked
```

```python
ranked[["content_id", "search_volume", "content_type", "impr_change_pct"]].head(10)
```

```python
(ranked["impr_change_pct"] == -1.0).sum()
```

```python
ranked = scoped.sort_values(["impr_change_pct", "impressions_prev_30d"], ascending=[True, False])
```

```python
ranked[["content_id", "search_volume", "impressions_prev_30d", "impr_change_pct"]].head(10)
```

```python
true_worst = set(ranked.head(100)["content_id"])
```

```python
naive_guess = set(scoped.sort_values("word_count").head(100)["content_id"])
```

```python
len(naive_guess & true_worst)
```

```python
import random
random_guess = set(random.sample(list(scoped["content_id"]), 100))
len(random_guess & true_worst)
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Why ML beats a fixed rule here: Individually, none of the obvious factors
(search volume, competition, word count, content age, days since last update)
correlate meaningfully with decline. A naive fixed rule ("shortest articles are
the worst decliners") scored 0/100 precision@100 - worse than a random baseline,
which scored 2/100. This shows decline isn't explained by any single factor, but
by combinations of weak signals interacting together - which is exactly the kind
of pattern ML models can learn and a fixed rule cannot.
